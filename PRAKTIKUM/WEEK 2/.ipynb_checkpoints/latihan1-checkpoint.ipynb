**Tahun Kabisat**

```python
tahun = int(input("Masukkan tahun:"))

kabisat = True

if (tahun % 4 == 0 and tahun % 100 != 0) or (tahun % 400 == 0):
    print(kabisat)
else:
    print(not kabisat)
```
