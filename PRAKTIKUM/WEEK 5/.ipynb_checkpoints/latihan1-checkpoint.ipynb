```python
import matplotlib.pyplot as plt
```

```python
hari = [1, 2, 3, 4, 5, 6, 7, 90]
toko_a = [12, 83, 23, 56, 26, 19, 33, 76]
toko_b = [33, 26, 11, 90, 31, 87, 71, 44]
```

```python
fig = plt.figure(figsize=(10,5))
```

```python
plt.scatter(hari, toko_a, c='blue', alpha=1, label="TOKO A")
plt.scatter(hari, toko_b, c='pink', alpha=0.65, label="TOKO B")
plt.title("BAJINGANNN")
plt.xlabel("Mokondo")
plt.ylabel("Jangan pake narkoba ya")
plt.legend()
plt.show()
```

```python
nama = ["Udin", "Bahar", "Safar", "Naura"]
nilai = [90, 20, 90, 100]
fig, ax = plt.subplots(figsize=(8,6))
list_color = ['red','maroon','black', 'grey']
ax.bar(nama, nilai, color = list_color)
ax.set_title('Jumlah Kendaraan yang Melewati 5 Ruas Jalan Utama dalam Sehari')
ax.set_xlabel('Ruas Jalan')
ax.set_ylabel('Jumlah Kendaraan')
plt.show()
```
