## Pengenalan Matplotlib
Matplotlib merupakan library Python yang digunakan untuk membuat **visualisasi data dalam bentuk grafik**. Library ini menyediakan berbagai alat untuk membuat grafik seperti **linechart**, **barchart**, **histogram**, **scatter plot**, dan banyak lagi.

````python
import matplotlib.pyplot as plt # Import library matplotlib.pyplot
````

### Soal 1: Visualisasi Scatter Plot

Sebuah pusat perbelanjaan ingin menganalisis pola penjualan dari dua toko selama satu minggu. Data berikut menunjukkan total transaksi harian (dalam ribuan rupiah) yang terjadi di dua toko tersebut.

Gunakan scatter plot untuk memvisualisasikan hubungan antara hari ke- dan total transaksi dari kedua toko.

| Hari ke- | Penjualan Toko A (ribu rupiah) | Penjualan Toko B (ribu rupiah) |
|---------|--------------------------|--------------------------|
| 1       | 200                      | 190                      |
| 2       | 220                      | 210                      |
| 3       | 200                      | 230                      |
| 4       | 220                      | 260                      |
| 5       | 270                      | 245                      |
| 6       | 320                      | 310                      |
| 7       | 290                      | 300                      |

````python
# Data pelanggan dan total penjualan (dalam ribuan rupiah)
hari = [1, 2, 3, 4, 5, 6, 7]
penjualan_toko_a = [200, 220, 200, 220, 270, 320, 290]
penjualan_toko_b = [190, 210, 230, 260, 245, 310, 300]

# Inisialisasi figure
fig = plt.figure(figsize=(10, 5))

# Plot scatterplot untuk data Toko A
plt.scatter(hari, penjualan_toko_a, c='red', alpha=0.65, label="Toko A")

# Plot scatterplot untuk data Toko B
plt.scatter(hari, penjualan_toko_b, c='blue', alpha=0.50, label="Toko B")

# Menambahkan judul dan label
plt.title("Hubungan Jumlah Pelanggan dan Total Penjualan di Dua Toko")
plt.xlabel("Hari ke-")
plt.ylabel("Total Penjualan (Ribu Rupiah)")

# Menambahkan legend
plt.legend()

# Tampilkan scatterplot
plt.show()
````

### Soal 2: Visualisasi Bar Chart

Dinas transportasi ingin mengetahui volume kendaraan yang melewati beberapa ruas jalan utama dalam sehari. Data berikut menunjukkan jumlah kendaraan yang melintasi 5 ruas jalan berbeda.

Gunakan bar chart untuk memvisualisasikan jumlah kendaraan yang melewati setiap ruas jalan.

| Ruas Jalan | Jumlah Kendaraan |
|------------|----------------|
| Jalan A    | 1100           |
| Jalan B    | 900            |
| Jalan C    | 1500           |
| Jalan D    | 1200           |
| Jalan E    | 1600           |

````python
# Data jumlah kendaraan di 5 ruas jalan
ruas_jalan = ["Jalan A", "Jalan B", "Jalan C", "Jalan D", "Jalan E"]
jumlah_kendaraan = [1100, 900, 1500, 1200, 1600]

# Inisialisasi figure
fig, ax= plt.subplots(figsize=(8, 6))

# List warna
list_color = ['red', 'blue', 'green', 'purple', 'orange']

# Plot barchart untuk data kendaraan
ax.bar(ruas_jalan, jumlah_kendaraan, color=list_color)

# Menambahkan judul dan label
ax.set_title('Jumlah Kendaraan yang Melewati 5 Ruas Jalan Utama dalam Sehari')
ax.set_xlabel('Ruas Jalan')
ax.set_ylabel('Jumlah Kendaraan')

# Tampilkan barchart
plt.show()
````

Contoh *output*:

![tp5_bar](./asset/tp5_bar.png)

### Soal 3: Visualisasi Histogram

Sebuah perusahaan riset ingin menganalisis distribusi waktu penggunaan aplikasi sosial media oleh sekelompok pengguna dalam sehari. Data berikut menunjukkan waktu (dalam jam) yang dihabiskan oleh berbagai pengguna aplikasi sosial media.

Gunakan histogram untuk memvisualisasikan distribusi waktu penggunaan aplikasi.

| Pengguna | Waktu Penggunaan (Jam) |
|----------|----------------------|
| 1        | 1                    |
| 2        | 2                    |
| 3        | 3                    |
| 4        | 2                    |
| 5        | 4                    |
| ...      | ...                  |
| 40       | 3                    |

````python
# Data waktu penggunaan aplikasi (dalam jam)
waktu_penggunaan = [1, 2, 3, 2, 4, 3, 5, 4, 2, 3, 4, 6, 5, 3, 2, 1, 6, 4, 2, 3,
                    5, 3, 2, 4, 5, 1, 6, 7, 3, 4, 2, 1, 3, 2, 4, 5, 3, 6, 4, 3]

# Inisialisasi figure
fig, ax = plt.subplots(figsize=(8, 6))

# Plot histogram waktu penggunaan aplikasi
ax.hist(waktu_penggunaan, bins=7, color='skyblue')

# Menambahkan judul dan label
ax.set_title('Distribusi Waktu Penggunaan Aplikasi Sosial Media dalam Sehari')
ax.set_xlabel('Waktu Penggunaan (Jam)')
ax.set_ylabel('Jumlah Pengguna')

# Tampilkan histogram
plt.show()
````

### Soal 4: Visualisasi Pie Chart

Sebuah restoran ingin mengetahui distribusi penjualan berbagai jenis makanan selama satu minggu. Data berikut menunjukkan jumlah pembelian dari masing-masing kategori makanan yang paling populer.

Gunakan pie chart untuk memvisualisasikan distribusi penjualan makanan di restoran tersebut.

| Kategori Makanan | Jumlah Pembelian |
|-----------------|----------------|
| Nasi Goreng     | 520            |
| Mie Ayam        | 430            |
| Soto            | 350            |
| Bakso           | 390            |
| Ayam Goreng     | 470            |

````python
# Data kategori makanan dan jumlah pembelian
kategori_makanan = ["Nasi Goreng", "Mie Ayam", "Soto", "Bakso", "Ayam Goreng"]
jumlah_pembelian = [520, 430, 350, 390, 470]

# Inisialisasi figure
fig, ax = plt.subplots(figsize=(8, 8))

# Plot piechart
ax.pie(jumlah_pembelian, labels=kategori_makanan, autopct='%1.1f%%')

# Menambahkan judul
ax.set_title('Distribusi Jenis Makanan yang Paling Sering Dibeli di Restoran')

# Tampilkan piechart
plt.show()
````

Contoh *output*:

![tp5_pie](./asset/tp5_pie.png)

### Soal 5: Visualisasi Boxplot

Sebuah layanan pengiriman makanan ingin menganalisis distribusi waktu penyelesaian pesanan oleh kurir dari tiga cabang restoran. Data berikut menunjukkan waktu penyelesaian pesanan (dalam menit) dari masing-masing cabang.

Gunakan boxplot untuk memvisualisasikan distribusi waktu penyelesaian pesanan oleh kurir.

| Cabang Restoran | Waktu Penyelesaian (Menit) |
|----------------|--------------------------|
| Cabang A      | 25, 30, 28, 35, 32, 31, 27, 29, 34, 33 |
| Cabang B      | 20, 22, 25, 21, 23, 26, 24, 22, 27, 28 |
| Cabang C      | 30, 35, 32, 31, 33, 36, 34, 30, 29, 37 |

````python
# sesuaikan data
cabang_a = [25, 30, 28, 35, 32, 31, 27, 29, 34, 33]
cabang_b = [20, 22, 25, 21, 23, 26, 24, 22, 27, 28]
cabang_c = [30, 35, 32, 31, 33, 36, 34, 30, 29, 37]

# Gabungkan data
data_nilai = [cabang_a, cabang_b, cabang_c]

# Boxplot
plt.figure(figsize=(8,6))
plt.boxplot(data_nilai, label=["Cabang A", "Cabang B", "Cabang C"], patch_artist=True)

# Tambahkan detail
plt.xlabel("Cabang")
plt.ylabel("Nilai Ujian")
plt.title("Distribusi Nilai Ujian per Cabang")
plt.grid(True)

# Tampilkan boxplot
plt.show()
````
